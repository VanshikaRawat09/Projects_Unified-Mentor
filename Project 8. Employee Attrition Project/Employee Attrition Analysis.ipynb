```python
# Check for missing values and handle them
df.dropna(inplace=True)  # Drop rows with missing values
```

```python
# Check for and remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Convert float data type columns to integer
float_columns = df.select_dtypes(include=['float']).columns
df[float_columns] = df[float_columns].astype(int)
```

```python
#Remove unwanted columns
df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1, inplace=True)
df
```

```python
# Analyze employee attrition
attrition_count = df['Attrition'].value_counts()
attrition_percentage = df['Attrition'].value_counts(normalize=True) * 100
```

```python
# Visualize the data
plt.figure(figsize=(6, 4))
plt.bar(attrition_count.index, attrition_count.values, color='skyblue')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.title('Employee Attrition Analysis')
for i, value in enumerate(attrition_count.values):
    plt.text(i, value + 1, str(value), ha='center')
plt.show()
```

```python
# Print the analysis results
print("Attrition Count:")
print(attrition_count)
print("\nAttrition Percentage:")
print(attrition_percentage)
```

```python
# Analyze other features in relation to the attrition column
# For example, let's analyze the Age distribution for employees with and without attrition
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
#Distribution of age
sns.displot(df['Age'], kde=True)
plt.title('Distribution of age')
plt.show()
```

```python
# Creating a count plot to visualize business travel by attrition
plt.figure(figsize=(8, 6))
sns.countplot(x='BusinessTravel', hue='Attrition', data=df)
plt.title('Business Travel by Attrition')
plt.xlabel('Business Travel')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate the percentage of attrition by department
attrition_percentage = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()['Yes'] * 100

# Plotting the donut chart
fig, ax = plt.subplots()
ax.pie(attrition_percentage, labels=attrition_percentage.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.title('Percentage of Attrition by Department')
plt.show()
```

```python
# Creating a box plot to visualize distance from home by attrition
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df)
plt.xlabel('Attrition')
plt.ylabel('Distance from Home')
plt.title('Distance from Home by Attrition')
plt.show()
```

```python
# Calculate the percentage of attrition by education field
attrition_percentage = df.groupby('EducationField')['Attrition'].value_counts(normalize=True).unstack().fillna(0) * 100

# Plotting the percentage of attrition by education field
attrition_percentage.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.xlabel('Education Field')
plt.ylabel('Percentage of Attrition')
plt.title('Percentage of Attrition by Education Field')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

```python
# Group the data by 'Gender' and 'Attrition' columns and count the occurrences
attrition_gender_counts = df.groupby(['Gender', 'Attrition']).size().unstack()

# Calculate the percentage of attrition for each gender
attrition_gender_counts['Attrition_Percentage'] = (attrition_gender_counts['Yes'] / attrition_gender_counts.sum(axis=1)) * 100

# Plotting the data using a pie chart
attrition_gender_counts['Attrition_Percentage'].plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Attrition by Gender')
plt.ylabel('')
plt.show()
```

```python
# Create a count plot to visualize attrition based on marital status
plt.figure(figsize=(10, 6))
sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
plt.title('Attrition based on Marital Status')
plt.show()
```

```python
# Visualizing job role by attrition
plt.figure(figsize=(10, 6))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.title('Job Role by Attrition')
plt.show()
```

```python
# Visualizing number of companies worked by attrition
plt.figure(figsize=(8, 5))
sns.countplot(x='NumCompaniesWorked', hue='Attrition', data=df)
plt.title('Number of Companies Worked by Attrition')
plt.show()
```

```python
# Calculate attrition percentage by numcompanies worked
attrition_by_numcompanies = df.groupby('NumCompaniesWorked')['Attrition'].value_counts(normalize=True).unstack()
#attrition_by_numcompanies = attrition_by_numcompanies.fillna(0)  # Fill NaN values with 0

# Plotting the data
attrition_by_numcompanies.plot(kind='bar', stacked=True)
plt.title('Attrition Percentage by Number of Companies Worked')
plt.xlabel('Number of Companies Worked')
plt.ylabel('Attrition Percentage')
plt.legend(title='Attrition')
plt.show()
```

```python
#Relationship between monthly income and the age of the employee
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['MonthlyIncome'], color='blue', alpha=0.5)
plt.title('Relationship between Monthly Income and Age')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.grid(True)
plt.show()
```

```python
#Salary hike wise attrition

salary_hike_groups = df.groupby('PercentSalaryHike')['Attrition'].value_counts().unstack()
salary_hike_groups.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Attrition based on Salary Hike Categories')
plt.xlabel('Percent Salary Hike')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition', loc='upper right')

plt.show()
```

```python
#Distribution of total working years by attrition
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['Attrition'] == 'Yes']['TotalWorkingYears'], df[df['Attrition'] == 'No']['TotalWorkingYears']], labels=['Attrition: Yes', 'Attrition: No'])
plt.title('Distribution of Total Working Years by Attrition')
plt.xlabel('Attrition Status')
plt.ylabel('Total Working Years')
plt.show()
```

```python
#Attrition based on training time last year
training_time_groups = df.groupby('TrainingTimesLastYear')['Attrition'].value_counts().unstack()
training_time_groups.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Attrition based on Training Time Last Year')
plt.xlabel('Training Times Last Year')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition', loc='upper right')

plt.show()
```

```python
# Create a bar chart to visualize attrition based on years at the company
years_at_company_groups = df.groupby('YearsAtCompany')['Attrition'].value_counts().unstack()
years_at_company_groups.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Attrition based on Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition', loc='upper right')

plt.show()
```

```python
# Create a bar chart to visualize attrition based on years since last promotion
years_since_last_promotion_groups = df.groupby('YearsSinceLastPromotion')['Attrition'].value_counts().unstack()
years_since_last_promotion_groups.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Attrition based on Years Since Last Promotion')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition', loc='upper right')

plt.show()
```

```python
# Create a bar chart to visualize attrition based on YearsWithCurrManager
years_with_curr_manager_groups = df.groupby('YearsWithCurrManager')['Attrition'].value_counts().unstack()
years_with_curr_manager_groups.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Attrition based on Years With Current Manager')
plt.xlabel('Years With Current Manager')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition', loc='upper right')

plt.show()
```

```python
# Filter the data for employees who have left (Attrition = 'Yes')
attrition_data = df[df['Attrition'] == 'Yes']

# Calculate the counts for EnvironmentSatisfaction, JobSatisfaction, and WorkLifeBalance
environment_satisfaction_counts = attrition_data['EnvironmentSatisfaction'].value_counts()
job_satisfaction_counts = attrition_data['JobSatisfaction'].value_counts()
work_life_balance_counts = attrition_data['WorkLifeBalance'].value_counts()

# Create a pie chart for EnvironmentSatisfaction
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.pie(environment_satisfaction_counts, labels=environment_satisfaction_counts.index, autopct='%1.1f%%')
plt.title('Environment Satisfaction for Employees who Left')

# Create a pie chart for JobSatisfaction
plt.subplot(1, 3, 2)
plt.pie(job_satisfaction_counts, labels=job_satisfaction_counts.index, autopct='%1.1f%%')
plt.title('Job Satisfaction for Employees who Left')

# Create a pie chart for WorkLifeBalance
plt.subplot(1, 3, 3)
plt.pie(work_life_balance_counts, labels=work_life_balance_counts.index, autopct='%1.1f%%')
plt.title('Work Life Balance for Employees who Left')

plt.tight_layout()
plt.show()
```

```python
# Selecting numerical columns for correlation matrix
numerical_columns = df.select_dtypes(include=['int64', 'float64'])

# Calculating the correlation matrix
correlation_matrix = numerical_columns.corr()

# Creating a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
